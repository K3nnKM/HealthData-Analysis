```python
# Set visual style
sns.set(style="whitegrid")

# 1. Heart Attack Risk distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Heart_Attack_Risk', palette='coolwarm')
plt.title('Heart Attack Risk Distribution')
plt.ylabel('Count')
plt.xlabel('Risk Level')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# Label Encoding for binary categorical columns
binary_cols = ['Gender', 'Smoking', 'Obesity', 'Diabetes', 'Family_History', 
               'Hypertension', 'Medication_Use', 'Previous_Heart_Attack']

le = LabelEncoder()
for col in binary_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Encode target variable: Heart_Attack_Risk
risk_mapping = {'Low': 0, 'Moderate': 1, 'High': 2}
df_encoded['Heart_Attack_Risk'] = df_encoded['Heart_Attack_Risk'].map(risk_mapping)

# One-Hot Encoding for nominal multi-class categorical columns
one_hot_cols = ['Alcohol_Consumption', 'Physical_Activity', 'Diet', 
                'Employment_Status', 'Income_Level', 'Marital_Status']

df_encoded = pd.get_dummies(df_encoded, columns=one_hot_cols, drop_first=True)

# Final check
print(df_encoded.shape)
df_encoded.head()
```

```python
numeric_features = ['Age', 'Blood_Pressure', 'Cholesterol', 
                    'Heart_Rate', 'Exercise_Hours_Per_Week', 'Stress_Levels']

for col in numeric_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, hue='Heart_Attack_Risk', kde=True, element='step', palette='coolwarm')
    plt.title(f'{col} Distribution by Heart Attack Risk')
    plt.show()
```

```python
categorical_cols = ['Smoking', 'Obesity', 'Hypertension', 'Diabetes', 'Diet', 
                    'Physical_Activity', 'Alcohol_Consumption']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, hue='Heart_Attack_Risk', palette='Set2')
    plt.title(f'{col} vs Heart Attack Risk')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

```python
# Encode binary columns before correlation
# Map binary and target values as before
df_corr = df.copy()

# Encode binary categorical columns
binary_cols = ['Gender', 'Smoking', 'Obesity', 'Diabetes', 'Family_History', 
               'Hypertension', 'Medication_Use', 'Previous_Heart_Attack']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in binary_cols:
    df_corr[col] = le.fit_transform(df_corr[col])

# Encode target variable
df_corr['Heart_Attack_Risk'] = df_corr['Heart_Attack_Risk'].map({'Low': 0, 'Moderate': 1, 'High': 2})

# Select only numeric columns
numeric_df = df_corr.select_dtypes(include=['int64', 'float64'])

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Start with a fresh copy
df_encoded = df.copy()

# Step 1: Label encode binary categorical variables
binary_cols = ['Gender', 'Smoking', 'Obesity', 'Diabetes', 'Family_History', 
               'Hypertension', 'Medication_Use', 'Previous_Heart_Attack']

le = LabelEncoder()
for col in binary_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Step 2: Encode target variable
df_encoded['Heart_Attack_Risk'] = df_encoded['Heart_Attack_Risk'].map({'Low': 0, 'Moderate': 1, 'High': 2})

# Step 3: One-hot encode multi-class categorical variables
multi_class_cols = ['Alcohol_Consumption', 'Physical_Activity', 'Diet', 
                    'Employment_Status', 'Income_Level', 'Marital_Status']

df_encoded = pd.get_dummies(df_encoded, columns=multi_class_cols, drop_first=True)

# Step 4: Drop ID column (not useful in correlation)
df_encoded = df_encoded.drop(columns=['Patient_ID'])

# Step 5: Plot correlation heatmap
plt.figure(figsize=(18, 14))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap (All Encoded Features)')
plt.show()
```
